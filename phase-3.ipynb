```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Data structure [restuarant_id, Q1_sales, Q2_sales, Q3_sales, Q4_sales]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 240000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])
print("Sales Data:")
print(sales_data.shape)
print('first 3 data ', sales_data[:3])
# print('Total sum', np.2ssum(sales_data, axis = 0))
year_total = np.sum(sales_data[:, 1:], axis=0)
print(year_total)

min_sales = np.min(sales_data[:, 1:], axis=1 )
print(min_sales)
#Maximum Sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print('max_sales', max_sales)

#Average
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)

# Cumulative sum
cum_sum = np.cumsum(sales_data[:, 1:], axis=1)
print(cum_sum)

plt.Figure(figsize=(10, 6))
plt.plot(np.mean(cum_sum, axis=0))
plt.title('Avg cumulative sales per year across all resturant')
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

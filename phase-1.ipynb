## NumPY Array and Basics

```python
import numpy as np
```

##  Creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print(arr_1d)

arr_2d = np.array([ [1, 2, 3], [4, 5, 6] ])
print(arr_2d)
```

## List V/S numpy Array

```python
py_list = [1,2,3,4]
print("python list multplication", py_list * 2) 

np_array = np.array([1,2,3,4])
print("print python array multiplication", np_array * 2) # element wise multiplications
```

## time

```python
import time 
start = time.time()
py_list = [ i * 2 for i in range(10000)]
print('\n ist operation', time.time() - start)

start = time.time()
np_array = np.arange(100000) * 2
print("\n Numpy operation time", time.time() - start)
```

## Array from scratch

```python
zeroes = np.zeros((3,4))
print(zeroes)

ones = np.ones( (2, 3))
print('once array', ones)

full = np.full( (2, 2), 7)
print("Full array", full)

random = np.random.random( (2, 3)) 
print("rando, array", random)

np_sequence = np.arange(0, 10, 2)
print("seqiqme array \n", np_sequence)

```

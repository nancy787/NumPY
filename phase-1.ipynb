## List V/S numpy Array

```python
py_list = [1,2,3,4]
print("python list multplication", py_list * 2) 

np_array = np.array([1,2,3,4])
print("print python array multiplication", np_array * 2) # element wise multiplications
```

## time

```python
import time 
start = time.time()
py_list = [ i * 2 for i in range(10000)]
print('\n ist operation', time.time() - start)

start = time.time()
np_array = np.arange(100000) * 2
print("\n Numpy operation time", time.time() - start)
```

## ARRAY RESHAPING

```python
arr = np.arange(12) #create array with 12 elemne star with 0
print("origital array", arr)

reshape = arr.reshape((3,4)) #3 rows and 4 column
print('Resshaped Aray', reshape)

flattened = reshape.flatten() #returns copy
print('flattened Aray', flattened)

raveled = reshape.ravel() #ravel re̋turns view instead of copy
print('Raveled array', raveled)

#Transpose˝ 
transpose = reshape.T
print('tranpose', transpose)
```
